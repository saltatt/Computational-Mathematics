```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.nn-airport.kz/flight-status', timeout= 10)

soup = BeautifulSoup (response.text, 'html.parser')
print(soup.prettify())
```

```python
tables = soup.find_all('table', class_='flight-status-table')
```

```python
well_find = soup.find('div',class_ = 'well')
table=well_find.find('table')
rows = table.find_all('tr')
flights = []
for row in rows:
            columns = row.find_all('td')

            if columns:
                flight_number = columns[0].get_text(strip=True)
                flight_status = columns[1].get_text(strip=True)
                flight_time = columns[2].get_text(strip=True)
                flight_gate = columns[3].get_text(strip=True)
                flights.append({
                    "flight_number": flight_number,
                    "status": flight_status,
                    "time": flight_time,
                    "gate": flight_gate
                })
df = pd.DataFrame(flights)
df
```

```python
df.to_csv('flights.csv', index=False)
```
