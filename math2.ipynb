```python
# Матрица коэффициентов A и вектор правых частей b
A = np.array([
    [14.1,  3.2,  5.1,  5.8],
    [6.6, -18.4,  8.2, -9.9],
    [9.3,  -7.6, 19.1, -3.8],
    [7.6,   5.5,  4.2, -14.7]
], dtype=float)

b = np.array([3.9, 7.2, 2.1, 9.1], dtype=float)

# Начальное приближение (нулевой вектор)
x = np.zeros_like(b)

# Параметры
eps = 1e-8
max_iter = 1000

print("Метод Гаусса–Зейделя")
print(f"Начальное приближение: x^(0) = {x}\n")

for k in range(1, max_iter + 1):
    x_old = x.copy()
    for i in range(len(b)):
        s1 = np.dot(A[i, :i], x[:i])        # уже обновлённые значения
        s2 = np.dot(A[i, i + 1:], x_old[i + 1:])  # старые значения
        x[i] = (b[i] - s1 - s2) / A[i, i]

    # Норма разности между итерациями
    diff = np.linalg.norm(x - x_old, ord=np.inf)

    print(f"Итерация {k:2d}: x = {x}, разность = {diff:.2e}")

    if diff < eps:
        print(f"\nСошлось за {k} итераций при ε = {eps}")
        break

# Проверка точности решения (остаток)
r = A @ x - b
residual = np.linalg.norm(r, ord=np.inf)

print("\nФинальное решение:")
for i, xi in enumerate(x, 1):
    print(f"x{i} = {xi:.10f}")

print(f"\nОстаток ||A·x - b||∞ = {residual:.2e}")
```
